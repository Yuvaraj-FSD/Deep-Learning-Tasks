```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import BatchNormalization, LayerNormalization
```

```python
# Create a random dataset
np.random.seed(42)

input_data = np.random.randint(1, 100, size=(5, 4)).astype(np.float32)

print("Original Input Values:")
print(input_data)
```

```python
# Convert NumPy array into TensorFlow tensor
input_tensor = tf.convert_to_tensor(input_data)

print("Tensor Shape:", input_tensor.shape)
```

```python
# Create Batch Normalization layer
batch_norm = BatchNormalization()

# Apply Batch Normalization
batch_output = batch_norm(input_tensor, training=True)

print("Batch Normalization Output:")
print(batch_output.numpy())
```

```python
# Create Layer Normalization layer
layer_norm = LayerNormalization()

# Apply Layer Normalization
layer_output = layer_norm(input_tensor)

print("Layer Normalization Output:")
print(layer_output.numpy())
```

```python
print("========== ORIGINAL INPUT ==========")
print(input_data)

print("\n========== BATCH NORMALIZATION OUTPUT ==========")
print(batch_output.numpy())

print("\n========== LAYER NORMALIZATION OUTPUT ==========")
print(layer_output.numpy())
```

```python
print("Batch Normalization:")
print("- Batch Normalization normalizes the input across the batch.")
print("- It helps speed up training and improves model stability.")

print("\nLayer Normalization:")
print("- Layer Normalization normalizes the features of each individual sample.")
print("- It works well for recurrent networks and small batch sizes.")

print("\nDifference:")
print("- Batch Normalization uses batch statistics, while Layer Normalization uses feature statistics for each sample.")
```
